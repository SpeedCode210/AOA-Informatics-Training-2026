# Big-O Notation

## What is Big-O Notation?

Big-O notation is a mathematical concept used in computer science to describe how the runtime or space requirements of an algorithm grow as the input size grows. It focuses on the **worst-case scenario** and provides an **upper bound** on the growth rate of an algorithm's complexity.

## Formal Definition

We say that a function $f(n) \in O(g(n))$ iff there exists a positive constant $c$ and a positive integer $n_0$ such that $\forall n > n_0, f(n) \leq cg(n)$.

## Why Do We Use Big-O Notation?

1. **Performance Comparison**: It allows us to compare algorithms objectively, regardless of hardware differences. Instead of saying "Algorithm A is faster than Algorithm B," we can say "Algorithm A has $O(n)$ complexity while Algorithm B has $O(n^2)$ complexity."

2. **Scalability Prediction**: Big-O helps predict how an algorithm will perform as the input size increases. An $O(n)$ algorithm will handle 10x more data in roughly 10x more time, while an $O(n^2)$ algorithm would take 100x more time.

3. **Code Optimization**: By understanding Big-O, developers can identify inefficient parts of code and focus optimization efforts where they'll have the most impact.

4. **Easier Computations**: Big-O notation allows us to simplify complex time and space functions into more manageable forms. Instead of dealing with precise equations like $T(n) = 3n^2 + 5n + 20$, we simplify it to $O(n^2)$. This simplification makes it much easier to perform computations and comparisons between different algorithms.


## Key Takeaways

Big-O notation ignores constants and lower-order terms, focusing only on the dominant term as input size approaches infinity. For example, O(2n + 3) simplifies to O(n).

It's not about exact timing but about understanding how algorithms scale with larger inputs. This knowledge helps developers choose the most efficient algorithms for their specific use cases.

## Exercise 3
Find an upper bound for the time complexity of the following function :

```c++
extern "C++"
int F(int n){
    if(n < 2) return 1;
    return F(n-1) + F(n-2);
}
```

```c++
int n = 10;
cout << F(n);
```

<details>
    <summary>Answer</summary>
    Let f(n) be the number of operations needed to compute Fₙ. Then:

$f(0) = f(1) = 1$

since it only needs one operation. For n larger:

$f(n) = f(n-1) + f(n-2)$

It is obvious that

$f(n-2) \leq f(n-1)$

Hence:

$f(n) \leq 2 × f(n-1)$

giving:

$f(n) \leq 2^n$

Therefore, we can consider our algorithm running in $O(2^n)$, which is an exponential complexity.

</details>

<details>
    <summary>Note</summary>
    Actually one can prove a better bound, which is that the function runs in $O(φ^n)$  where  $φ≈1.618$  (the golden ratio), but this is not relevant for our lesson.
</details>

## **Time Complexity Cheat Sheet (For Early CP Practice)**

⚠️ **Important Note:**
*Don’t memorize this table!* As you solve more problems, you’ll naturally recognize which complexities work best. This is just a starting point to help you estimate performance early in your competitive programming (CP) journey.

| **Time Complexity** | **Max Input Size (n)** |
|---------------------|-----------------------|
| $O(\log n)$         | Very large (e.g., $10^{18}$) |
| $O(\sqrt{n})$       | $n \leq 10^{14}$      |
| $O(n)$              | $n \leq 10^7$         |
| $O(n \log n)$       | $n \leq 10^6$         | 
| $O(n^2)$            | $n \leq 5000$         |
| $O(n^3)$            | $n \leq 500$          |
| $O(2^n)$            | $n \leq 20$           |
| $O(n!)$             | $n \leq 10$           | 

🔹 **Remember:**
- Time complexity is an **estimate**—real-world performance depends on constant factors.
- If your solution is close to the limit (e.g., $5 \times 10^7$ operations), it *might* still pass!
- **Advice:** If your code is near the boundary, submit it anyway—you might get lucky!

### **How Fast is My Algorithm? (A Practical Guide)**

Ever wondered what that **"3.70 GHz"** on your CPU means?

That number tells you how many **operations per second** your CPU can handle:
- **3.70 GHz = 3.7 billion operations per second** (theoretical max).

But here’s the catch:
- **C++ isn’t executed directly by the CPU**—there’s overhead.
- **Big-O ignores constants**, so real-world performance varies.

This table helps you **estimate** whether your algorithm will run in time for a given input size.

If your solution is borderline, **submit it anyway!** Many problems have hidden optimizations, and sometimes the judge’s hardware is faster than expected, or the testcases weaker than expected.



## On the $O(\log n)$ Complexity

Most likely, you will encounter this complexity when dealing with algorithms that divide the problem into two parts each time, running approximately $C \times \log_2 n$ operations.

**Example:** Suppose that you have a piece of cake 🍰 of length $n$ centimeters, and you put it on a table. Then you keep doing the following operation: Take the leftmost piece of cake on the table and cut it equally into two parts, until the leftmost piece is less than 1 cm long. Hence, you will have to do around $\log_2 n$ operations.

For $n = 4$: $4 \rightarrow 2 \rightarrow 1 \rightarrow 0.5$

For $n = 12$: $12 \rightarrow 6 \rightarrow 3 \rightarrow 1.5 \rightarrow 0.75$

For $n = 16$: $16 \rightarrow 8 \rightarrow 4 \rightarrow 2 \rightarrow 1 \rightarrow 0.5$

You will generally find this complexity when dealing with the paradigm of Divide-and-Conquer, sortings, and binary search. I won't spoil the fun here, haha!
